```python
import pandas as pd
import seaborn as sns
import pandas.testing as pd_testing
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Boston.csv")
df.head()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle
```

```python
x_train,X_test,y_train,y_test = train_test_split(df[['CRIM ','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']],df['PRICE'],test_size=0.3,random_state=100)
```

```python
linearRegr = LinearRegression()
linearRegr.fit(x_train,y_train)
y_pred = linearRegr.predict(X_test)
```

```python
pickle_out = open("linearRegr_multi1.pkl","wb")
pickle.dump(linearRegr,pickle_out)
pickle_out.close()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,y_pred)
print("Mean Squared Error: ",mse)

r2 = r2_score(y_test,y_pred)
print("R-squared: ",r2)
```
